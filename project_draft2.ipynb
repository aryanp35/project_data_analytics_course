```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import glob
import math
import numpy as np
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=4)
```

```python
cd "D:\projectfiles\Training Dataset"
```

```python
path = r'D:\projectfiles\Training Dataset\character_0'# use your path
all_files = glob.glob(path + "/*.csv")
X = []
y = []
for i in range(10):
    for j in range(1,1001):
        path = 'D:\projectfiles\Training Dataset/character_' + str(i) + '/' + str(j) + '.csv'
        img = np.loadtxt(path, delimiter=',')
        img = imputer.fit_transform(img).T
        x = img.reshape(1,784)[0]
        X.append(x)
        y.append(i)
```

```python
data = pd.DataFrame(X)
data['label'] = y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(['label'],axis=1)
X=X/255.0
y = data['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=40,whiten=True).fit(X_train)
X_train_pca = pca.transform(X_train)
```
